```python
#relational operator
x = 10
y = 5
print(x > y)
print(x < y)
print(x == y)
print(x != y)
```

```python
#assignment operator
a = 5
b = 10
a += 3
b -= 4
print(a)
print(b)
```

```python
#arithmetic operators
c = 7
d = 3
print(c + d)
print(c - d)
print(c / d)
print(c // d)
print(c % d)
```
